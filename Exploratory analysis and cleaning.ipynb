```python
missing_values.head(20)
```

```python
plt.figure(figsize = (20,14))
missing_values.plot(kind = 'bar')
plt.show()
```

```python
missing = missing_values[missing_values > 0]
```

```python
plt.figure(figsize = (20,14))
missing.plot(kind = 'bar')
plt.show()
```

```python
print("Number of features with missing data in train : ", len(missing))
```
